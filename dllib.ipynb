### feed data

```python
def build_model(input_shape=28*28, classes=10, hidden_size=64, l=0.005):
    l_in = Input(dim=input_shape)
    l_y = Input(dim=classes)

    l_h1 = Dense(l_in, dim_out=hidden_size)
    l_h1_a = Sigmoid(l_h1)
    l_h15 = Dense(l_h1_a, dim_out=hidden_size)
    l_h15_a = Sigmoid(l_h15)
    l_h2 = Dense(l_h15_a, dim_out=classes)
    l_h2_a = Sigmoid(l_h2)
    
    model = Model(l_in, l_h2_a, l_y, MeanSquaredError, SGD())
    
    return model
```

```python
m = build_model(hidden_size=20)
```

```python
from sklearn import datasets
mnist = datasets.fetch_openml('mnist_784', data_home='.')
```

```python
ins = mnist.data
labels = mnist.target

num_train = 30000
num_test = 500


X_tr = ins[:num_train] / 256
y_tr = np.eye(10)[[int(x) for x in labels[:num_train]]]
X_te = ins[num_train:num_train+num_test] / 256
y_te = np.eye(10)[[int(x) for x in labels[num_train:num_train+num_test]]]
```

```python
idx = 0
m.fit(X_tr, y_tr, epochs=5)
```

```python
conf = [[0 for _ in range(10)] for _ in range(10)]
for X, y in zip(X_te, y_te):
    pred = m.predict(X)
    
    idx_p = np.argmax(pred)
    idx_a = np.argmax(y)
    
    conf[idx_a][idx_p] += 1
```
